```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

### Load the data you receive into a Pandas DataFrame.


```python
df= pd.read_csv('./data.csv')
df
```

### Show the first five rows of the data set.

```python
first_five_rows=df.head(5)
first_five_rows
```

```python
y=df['Monthly_Rent'].values
y
```

```python
X=df['Shop_Area'].values
X
```

### Using a regression model, split your data into train and test data and fit your training split to the regression model.


```python
X=X.reshape(-1,1)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
print(f"X_train shape {X_train.shape}")
print(f"y_train shape {y_train.shape}")
print(f"X_test shape {X_test.shape}")
print(f"y_test shape {y_test.shape}")
plt.scatter(df['Shop_Area'],df['Monthly_Rent'],color='blue')
plt.xlabel('Shop Space')
plt.ylabel('Rent')
```

```python
plt.scatter(X_train,y_train,color='green')
plt.xlabel('Shop_Area')
plt.ylabel('Monthly_Rent')

plt.title('Monthly_Rent vs Shop_Area')
plt.show()
```

### Show your regression models score.


```python
model = LinearRegression()
model.fit(X_train,y_train)
print(f"Train accuracy {round(model.score(X_train,y_train)*100,2)} %")
print(f"Test accuracy {round(model.score(X_test,y_test)*100,2)} %")
```

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, model.predict(X_train), color = 'blue')
plt.title('Shop_Area vs Monthly_Rent (Training set)')
plt.xlabel('Shop_Area')
plt.ylabel('Monthly_Rent')
plt.show()
```

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, model.predict(X_train), color = 'blue')
plt.title('Shop_Area vs Monthly_Rent (Test set)')
plt.xlabel('Shop_Area')
plt.ylabel('Monthly_Rent')
plt.show()
```

```python

from sklearn.metrics import r2_score
r2_score(y_test, model.predict(X_test))*100
```
